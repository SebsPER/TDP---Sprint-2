```python
import tensorflow_decision_forests as tfdf
import pandas as pd
import random
import numpy as np
import tensorflow as tf
```

```python
url = "https://raw.githubusercontent.com/lmcanavals/ml/main/data/iris.data"
```

```python
data = pd.read_csv(url, header=None)
data.rename(columns={0: 'data1', 1: 'data2', 2:'data3', 3:'data4', 4: 'target'}, inplace = True)
```

```python
data.head()
```

```python
test_df = data.drop(['target'], axis=1)
test_df.head()
```

```python
# Convert the pandas dataframe into a TensorFlow dataset
train_ds = tfdf.keras.pd_dataframe_to_tf_dataset(data, label="target")
test_ds = tfdf.keras.pd_dataframe_to_tf_dataset(test_df)
```

```python
# Train the model
model = tfdf.keras.RandomForestModel()
model.fit(x=train_ds)
```

```python
tfdf.model_plotter.plot_model_in_colab(model, tree_idx=0)
```

```python
data.iloc[0][:4]
```

```python
preds = model.predict(test_ds) ###{'data1':5,'data2':4,'data3':7,'data4':2}
#print(preds[0:5])
np.random.shuffle(preds)
print(preds[0:10])
```
